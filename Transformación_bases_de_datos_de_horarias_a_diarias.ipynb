```python
# Datos meteorol√≥gicos
df_meteorologicas = pd.read_csv(
    "Meteorologicas_220002_01012019_01072025_UTC.csv",
    dayfirst=True                         # si usas formato d√≠a/mes/a√±o
)
```

```python
import pandas as pd

# 2) Eliminar la columna 'codigoNacional.x' si existe
if "codigoNacional.x" in df_meteorologicas.columns:
    df_meteorologicas.drop(columns=["codigoNacional.x"], inplace=True)

# 2) Renombrar columnas
rename_dict = {
    "momento":     "fecha_hora",
    "isSkyClear":  "cielo_despejado_proporcion",
    "hr":          "humedad_promedio",
    "qff":         "presion_promedio",
    "morInst":     "rango_optico_meteorologico_promedio",
    "ts":          "temperatura_aire_promedio",
    "ddInst":      "direccion_viento_promedio",
    "ffInst":      "intensidad_viento_promedio",
    "vis1Minuto":  "visibilidad_vertical_promedio"
}
df_meteorologicas.rename(columns=rename_dict, inplace=True)

```

```python
import pandas as pd

# 0) Asegurar que df_meteorologicas est√° indexado por datetime
df_m = df_meteorologicas.copy()
if not isinstance(df_m.index, pd.DatetimeIndex):
    if 'fecha_hora' in df_m.columns:
        df_m['fecha_hora'] = pd.to_datetime(df_m['fecha_hora'])
        df_m.set_index('fecha_hora', inplace=True)
    else:
        df_m.index = pd.to_datetime(df_m.index)
df_m.sort_index(inplace=True)

# 0.1) Filtrar para ignorar registros de las 07:00 y las 21:00
df_m = df_m[~df_m.index.hour.isin([7, 21])]

# 1) Agregar de horario a diario
daily = df_m.resample('D').agg({
    # Para temperatura: m√°ximo, promedio y m√≠nimo diario
    'temperatura_aire_promedio': ['max', 'mean', 'min'],
    # Para las dem√°s variables: promedio diario
    'cielo_despejado_proporcion':    'mean',
    'humedad_promedio':               'mean',
    'presion_promedio':               'mean',
    'rango_optico_meteorologico_promedio': 'mean',
    'direccion_viento_promedio':      'mean',
    'intensidad_viento_promedio':     'mean',
    'visibilidad_vertical_promedio':  'mean'
})

# 2) Aplanar MultiIndex de columnas y renombrar
daily.columns = [
    'temperatura_maxima',
    'temperatura_promedio',
    'temperatura_minima',
    'cielo_despejado_proporcion',
    'humedad_promedio',
    'presion_promedio',
    'rango_optico_meteorologico_promedio',
    'direccion_viento_promedio',
    'intensidad_viento_promedio',
    'visibilidad_vertical_promedio'
]

# 3) Resultado
print(daily.head())

# Si lo quieres en tu variable final:
df_meteorologicas_day = daily
```

```python
import pandas as pd

# 0) Asegurar que df_meteorologicas est√° indexado por datetime
if not isinstance(df_meteorologicas.index, pd.DatetimeIndex):
    if 'fecha_hora' in df_meteorologicas.columns:
        df_meteorologicas['fecha_hora'] = pd.to_datetime(df_meteorologicas['fecha_hora'])
        df_meteorologicas.set_index('fecha_hora', inplace=True)
    else:
        df_meteorologicas.index = pd.to_datetime(df_meteorologicas.index)
df_meteorologicas.sort_index(inplace=True)

# 1) Agregar de horario a diario
daily = df_meteorologicas.resample('D').agg({
    # Para temperatura: m√°ximo, promedio y m√≠nimo diario
    'temperatura_aire_promedio': ['max', 'mean', 'min'],
    # Para las dem√°s variables: promedio diario
    'cielo_despejado_proporcion': 'mean',
    'humedad_promedio': 'mean',
    'presion_promedio': 'mean',
    'rango_optico_meteorologico_promedio': 'mean',
    'direccion_viento_promedio': 'mean',
    'intensidad_viento_promedio': 'mean',
    'visibilidad_vertical_promedio': 'mean'
})

# 2) Aplanar MultiIndex de columnas y renombrar
daily.columns = [
    'temperatura_maxima',
    'temperatura_promedio',
    'temperatura_minima',
    'cielo_despejado_proporcion',
    'humedad_promedio',
    'presion_promedio',
    'rango_optico_meteorologico_promedio',
    'direccion_viento_promedio',
    'intensidad_viento_promedio',
    'visibilidad_vertical_promedio'
]

# 3) Verificar resultado
print(daily.head())
df_meteorologicas_day = daily
```

```python
# 1) Asegurar que el √≠ndice es DatetimeIndex
df_meteorologicas_day.index = pd.to_datetime(df_meteorologicas_day.index)

# 2) Renombrar el √≠ndice
df_meteorologicas_day.index.rename('fecha_dia', inplace=True)
```

```python
import pandas as pd

# 1) Resetear √≠ndice para tener la fecha como columna
df_export = df_meteorologicas_day.reset_index()

# 2) Asegurarte de que 'fecha_dia' no tenga zona horaria
# Si viene con tz (p.ej. America/Santiago), qu√≠tala:
df_export['fecha_dia'] = pd.to_datetime(df_export['fecha_dia']).dt.tz_localize(None)

# 3) Exportar a Excel con openpyxl o xlsxwriter
# Aseg√∫rate de haber instalado: pip install openpyxl
df_export.to_excel('Meteorologicas_220002_01012019_01072025_UTC.xlsx', index=False, engine='openpyxl')
```

```python
# Datos de generacion de energ√≠a de planta solar
df_solar = pd.read_csv(
    "Generacion_Energia_Solar_Sol_Desierto_01012024_02072025_UTC.csv",
    dayfirst=True                # si tus fechas vienen en formato d√≠a/mes/a√±o
)
```

```python
# 2) Renombrar las columnas seg√∫n especificaci√≥n
rename_dict = {
    "datetime":                     "fecha_hora",
    "PFV SOL DEL DESIERTO":         "generacion_MHw_sol_desierto"
}
df_solar.rename(columns=rename_dict, inplace=True)

# 3) Fijar 'fecha_hora' como √≠ndice datetime
df_solar.set_index("fecha_hora", inplace=True)
```

```python
import pandas as pd

# 0) Asegurar que df_solar est√° indexado por datetime
if not isinstance(df_solar.index, pd.DatetimeIndex):
    if 'fecha_hora' in df_solar.columns:
        df_solar['fecha_hora'] = pd.to_datetime(df_solar['fecha_hora'])
        df_solar.set_index('fecha_hora', inplace=True)
    else:
        df_solar.index = pd.to_datetime(df_solar.index)
df_solar.sort_index(inplace=True)

# 1) Agregar generaci√≥n horaria a diaria acumulada
daily_solar = df_solar['generacion_MHw_sol_desierto'] \
    .resample('D') \
    .sum() \
    .to_frame('generacion_MWh_diaria')

# 2) Verificar resultado
print(daily_solar.head())
df_solar_day = daily_solar
```

```python
import pandas as pd

# 1) Resetear √≠ndice para tener la fecha como columna
df_export = df_solar_day.reset_index()

# 2) Asegurarte de que 'fecha_dia' no tenga zona horaria
# Si viene con tz (p.ej. America/Santiago), qu√≠tala:
df_export['fecha_hora'] = pd.to_datetime(df_export['fecha_hora']).dt.tz_localize(None)

# 3) Exportar a Excel con openpyxl o xlsxwriter
# Aseg√∫rate de haber instalado: pip install openpyxl
df_export.to_excel('Generacion_Energia_Solar_Sol_Desierto_01012024_02072025_UTC.xlsx', index=False, engine='openpyxl')
```
